```python
# Dictionary to store output
outputs = {}

# Hook function
def hook(module, input, output):
    outputs['stage4'] = output

# Register hook to the correct layer
model.features[2].register_forward_hook(hook)

# Dummy input (B, C, H, W) = (1, 3, 300, 300)
x = torch.randn(1, 3, 300, 300)
with torch.no_grad():
    _ = model(x)

# Print output shape
print("Output after Stage 4:", outputs['stage4'].shape)
```
